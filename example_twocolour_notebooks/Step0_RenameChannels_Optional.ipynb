```python
import os

folder = os.path.abspath(
    r"../example_images"
)  # folder to recursively rename channel files in, replace as you wish.
   # Always input as the absolute path of a raw string

# Declare your channel-string renames here. Keys are the current substrings in
# your filenames (e.g. "C0"), values are what you want them renamed to (e.g.
# a wavelength-based label like "C640"). Add, remove, or edit entries as
# needed -- e.g. delete the "C2" line below if you don't have a third
# (bulk stain) channel, or add further entries (e.g. "C3": "C405") if you do.
channel_rename_map = {
    "C0": "C640",
    "C1": "C561",
    "C2": "C488",
}

dry_run = True  # keep True to preview renames first; set False to actually rename
```

```python
"""Recursively walks `folder` (including all subfolders) and renames every file
whose name contains one of the old channel strings in `channel_rename_map`
(e.g. "C0") to use the new string instead (e.g. "C640"). Only the filename is
changed, never any part of the folder path.

With dry_run = True (default), this only prints a preview of what would be
renamed -- no files are touched. Set dry_run = False above and re-run this
cell once you're happy with the preview.

A rename is skipped (and reported) if a file with the target name already
exists, so this will never silently overwrite an existing file.
"""


def planned_renames(folder, channel_rename_map):
    renames = []
    for dirpath, _, filenames in os.walk(folder):
        for fname in filenames:
            new_fname = fname
            for old, new in channel_rename_map.items():
                new_fname = new_fname.replace(old, new)
            if new_fname != fname:
                renames.append((dirpath, fname, new_fname))
    return renames


renames = planned_renames(folder, channel_rename_map)

print(f"{len(renames)} file(s) would be renamed:")
for dirpath, fname, new_fname in renames:
    print(f"  {os.path.join(dirpath, fname)}  ->  {new_fname}")

if not dry_run:
    n_done, n_skipped = 0, 0
    for dirpath, fname, new_fname in renames:
        src = os.path.join(dirpath, fname)
        dst = os.path.join(dirpath, new_fname)
        if os.path.exists(dst):
            print(f"SKIPPED (target already exists): {dst}")
            n_skipped += 1
            continue
        os.rename(src, dst)
        n_done += 1
    print(f"\nRenamed {n_done} file(s); skipped {n_skipped} due to name collisions.")
else:
    print(
        "\ndry_run is True -- no files were renamed. "
        "Set dry_run = False above and re-run to apply."
    )
```

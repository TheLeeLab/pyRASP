```python
overall_folder = r"/scratch/sycamore-asap/ASAP_Imaging_Data/Main_Survey/20241105_oligomers_in_cells/caudate/neurons"
old_raw_cells = H_F.file_search(overall_folder, "C0.tif", "")
old_cells = H_F.file_search(overall_folder, "cellMask.tiff", "")
new_cells = H_F.file_search(overall_folder, "cellMask_cleanedup.tiff", "")
```

```python
fig, axs = plotter.two_column_plot(
    ncolumns=2, nrows=3, widthratio=[1, 1], heightratio=[1, 1, 1]
)

i = 3
raw_data = np.max(IO.read_tiff(old_raw_cells[i]), axis=0)
image_nonclean = np.max(IO.read_tiff(old_cells[i]), axis=0).clip(0, 1)
image_clean = np.max(IO.read_tiff(new_cells[i]), axis=0).clip(0, 1)

axs[0] = plotter.image_plot(axs[0], raw_data, cbar="off")
axs[0].contour(image_nonclean, [0.5], linewidths=0.5, colors="red")

axs[1] = plotter.image_plot(axs[1], raw_data, cbar="off")
axs[1].contour(image_clean, [0.5], linewidths=0.5, colors="white")


i = 5
raw_data = np.max(IO.read_tiff(old_raw_cells[i]), axis=0)
image_nonclean = np.max(IO.read_tiff(old_cells[i]), axis=0).clip(0, 1)
image_clean = np.max(IO.read_tiff(new_cells[i]), axis=0).clip(0, 1)

axs[2] = plotter.image_plot(axs[2], raw_data, cbar="off")
axs[2].contour(image_nonclean, [0.5], linewidths=0.5, colors="red")

axs[3] = plotter.image_plot(axs[3], raw_data, cbar="off")
axs[3].contour(image_clean, [0.5], linewidths=0.5, colors="white")


i = 13
raw_data = np.max(IO.read_tiff(old_raw_cells[i]), axis=0)
image_nonclean = np.max(IO.read_tiff(old_cells[i]), axis=0).clip(0, 1)
image_clean = np.max(IO.read_tiff(new_cells[i]), axis=0).clip(0, 1)

axs[4] = plotter.image_plot(axs[4], raw_data, cbar="off")
axs[4].contour(image_nonclean, [0.5], linewidths=0.5, colors="red")

axs[5] = plotter.image_plot(axs[5], raw_data, cbar="off")
axs[5].contour(image_clean, [0.5], linewidths=0.5, colors="white")

plt.tight_layout()

folder = "/scratch/sycamore-asap/ASAP_Imaging_Data/Main_Survey"
plt.savefig(os.path.join(folder, "Thresholded_Cells_MIP.svg"), dpi=1200, format="svg")
```

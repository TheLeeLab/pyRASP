```python
import re                                                                                                                                                                                                                                                      
from pathlib import Path                                                                                                                                                                                                                                       
                                                                                                                                                                                                                                                             
                                                        
def _parse_renames_from_log(log_path):
  """Parse one log file into an explicit TIF rename dict and channel map.
                                                                                                                                                                                                                                                             
  Returns:
      tif_renames:  {old_tif_filename: new_tif_filename} for every row                                                                                                                                                                                       
                    listed in this log.                                                                                                                                                                                                                      
      channel_map:  {channel_index: laser_str} discovered from this log.                                                                                                                                                                                     
  """                                                                                                                                                                                                                                                        
  with open(log_path, 'r') as fh:                                                                                                                                                                                                                            
      lines = fh.readlines()                            
                                                                                                                                                                                                                                                             
  header_idx = next(
      (i for i, l in enumerate(lines) if l.startswith('IFD\t')), None                                                                                                                                                                                        
  )                                                     
  if header_idx is None:
      return {}, {}

  cols     = lines[header_idx].strip().split('\t')                                                                                                                                                                                                           
  ch_col   = cols.index('Channel Name')
  file_col = cols.index('TIFF File Name')                                                                                                                                                                                                                    
                                                        
  tif_renames = {}
  channel_map = {}

  for line in lines[header_idx + 1:]:                                                                                                                                                                                                                        
      row = line.strip().split('\t')
      if len(row) <= max(ch_col, file_col):                                                                                                                                                                                                                  
          continue                                      
      old_name = row[file_col].strip()
      if not old_name:
          continue
      m_idx   = re.search(r'(_C)(\d+)(\.tif)$', old_name)                                                                                                                                                                                                    
      m_laser = re.match(r'(\d+)', row[ch_col].strip())
      if not (m_idx and m_laser):                                                                                                                                                                                                                            
          continue                                                                                                                                                                                                                                           
      ch_idx = int(m_idx.group(2))
      laser  = m_laser.group(1)                                                                                                                                                                                                                              
      channel_map[ch_idx] = laser                       
      new_name = old_name[:m_idx.start()] + f'_C{laser}' + m_idx.group(3)
      if new_name != old_name:                                                                                                                                                                                                                               
          tif_renames[old_name] = new_name
                                                                                                                                                                                                                                                             
  return tif_renames, channel_map                                                                                                                                                                                                                            

                                                                                                                                                                                                                                                             
def rename_channels_recursive(root, dry_run=True):        
  """Rename _C0/_C1 → _C561/_C488 etc., driven by what each log explicitly lists.
                                                                                                                                                                                                                                                             
  For each directory:
    - Every .log file is parsed; the TIF filenames it names are added to the                                                                                                                                                                                 
      rename plan.  Together the logs in a folder should cover all TIFs.                                                                                                                                                                                     
    - Only files explicitly named in a log are renamed; anything else is left                                                                                                                                                                                
      untouched.                                                                                                                                                                                                                                             
    - .log files themselves are renamed using the accumulated channel map.                                                                                                                                                                                   
    - A warning is printed if two logs in the same folder disagree on a rename.                                                                                                                                                                              
    - Missing files (named in a log but absent on disk) are reported.                                                                                                                                                                                        
                                                                                                                                                                                                                                                             
  Args:                                                                                                                                                                                                                                                      
      root:    Root folder to walk recursively (str or Path).
      dry_run: If True (default) print planned renames without touching files.                                                                                                                                                                               
  """                                                                                                                                                                                                                                                        
  root = Path(root)                                                                                                                                                                                                                                          
  dirs_with_logs = sorted({p.parent for p in root.rglob('*.log')})                                                                                                                                                                                           
  total_renamed  = 0                                    

  for folder in dirs_with_logs:                                                                                                                                                                                                                              
      all_tif_renames = {}   # old_name → new_name, accumulated across all logs
      channel_map     = {}   # ch_idx   → laser_str                                                                                                                                                                                                          
                                                        
      for log in sorted(folder.glob('*.log')):                                                                                                                                                                                                               
          tif_renames, this_map = _parse_renames_from_log(log)
          if not tif_renames and not this_map:                                                                                                                                                                                                               
              continue
                                                                                                                                                                                                                                                             
          # Warn on conflicting renames from different logs in the same folder                                                                                                                                                                               
          for old, new in tif_renames.items():
              if old in all_tif_renames and all_tif_renames[old] != new:                                                                                                                                                                                     
                  print(f'  [WARN] Conflicting rename for {old}:\n'                                                                                                                                                                                          
                        f'         already={all_tif_renames[old]}, '
                        f'from {log.name}={new}')                                                                                                                                                                                                            
          all_tif_renames.update(tif_renames)           
          channel_map.update(this_map)                                                                                                                                                                                                                       
                                                        
      if not all_tif_renames and not channel_map:                                                                                                                                                                                                            
          continue                                      

      print(f'\n{folder}')
      print(f'  channel map: {channel_map}')
                                                                                                                                                                                                                                                             
      # --- Rename TIFs (only those explicitly named in a log) ---
      for old_name, new_name in sorted(all_tif_renames.items()):                                                                                                                                                                                             
          old_path = folder / old_name                                                                                                                                                                                                                       
          new_path = folder / new_name
          if not old_path.exists():                                                                                                                                                                                                                          
              print(f'  [MISS] {old_name}')             
              continue
          print(f'  {"[DRY]" if dry_run else "[OK] "} {old_name}  →  {new_name}')
          if not dry_run:                                                                                                                                                                                                                                    
              old_path.rename(new_path)
          total_renamed += 1                                                                                                                                                                                                                                 
                                                                                                                                                                                                                                                             
      # --- Rename log files using the channel map ---
      for log in sorted(folder.glob('*.log')):                                                                                                                                                                                                               
          m = re.search(r'(_C)(\d+)(\.log)$', log.name) 
          if not m:
              continue                                                                                                                                                                                                                                       
          ch_idx = int(m.group(2))
          if ch_idx not in channel_map:                                                                                                                                                                                                                      
              continue                                  
          new_name = log.name[:m.start()] + f'_C{channel_map[ch_idx]}' + m.group(3)
          new_path = folder / new_name                                                                                                                                                                                                                       
          if new_path == log:
              continue                                                                                                                                                                                                                                       
          print(f'  {"[DRY]" if dry_run else "[OK] "} {log.name}  →  {new_name}')
          if not dry_run:                                                                                                                                                                                                                                    
              log.rename(new_path)
          total_renamed += 1                                                                                                                                                                                                                                 
                                                        
  action = 'Would rename' if dry_run else 'Renamed'
  print(f'\n{action} {total_renamed} files across {len(dirs_with_logs)} '
        f'director{"y" if len(dirs_with_logs) == 1 else "ies"}.')                                                                                                                                                                                            
```

```python
import os                                                             
import re                                                                                                                                                                                                                                                      
import pandas as pd                                                                                                                                                                                                                                            
from pathlib import Path                                                                                                                                                                                                                                       
                                                                                                                                                                                                                                                             
                                                                                                                                                                                                                                                             
def collate_spot_analysis(root_dir, output_path="collated_data.h5"):
  """                                                                                                                                                                                                                                                        
  Recursively finds all spot_analysis.csv files under root_dir,
  tags each row with patient_id and group (HC or PD), and saves
  the collated DataFrame to an HDF5 file.                                                                                                                                                                                                                    
                                                                                                                                                                                                                                                             
  HC: folder name starts with 'C' or 'PDC'                                                                                                                                                                                                                   
  PD: folder name starts with 'PD' (but not 'PDC')                                                                                                                                                                                                           
                                                                                                                                                                                                                                                             
  Patient ID extraction (from the immediate parent folder name):
    - If folder contains '-': take the substring before the first '-'                                                                                                                                                                                        
    - Otherwise: take the substring before the first '_'                                                                                                                                                                                                     

  Parameters                                                                                                                                                                                                                                                 
  ----------  
  root_dir : str or Path                                                                                                                                                                                                                                     
      Top-level directory to search recursively.
  output_path : str or Path
      Path for the output .h5 file.                                                                                                                                                                                                                          

  Returns                                                                                                                                                                                                                                                    
  -------     
  pd.DataFrame
      The collated DataFrame (also saved to output_path).
  """                                                                                                                                                                                                                                                        
  root_dir = Path(root_dir)
  records = []                                                                                                                                                                                                                                               
                                                                                                                                                                                                                                                             
  for csv_path in sorted(root_dir.rglob("spot_analysis.csv")):
      folder_name = csv_path.parent.name                                                                                                                                                                                                                     
                                                                                                                                                                                                                                                             
      # --- determine patient_id ---
      if "-" in folder_name:                                                                                                                                                                                                                                 
          patient_id = folder_name.split("-")[0]
      elif "_" in folder_name:
          patient_id = folder_name.split("_")[0]                                                                                                                                                                                                             
      else:
          patient_id = folder_name                                                                                                                                                                                                                           
              
      # --- determine group ---
      if re.match(r"^PDC", folder_name, re.IGNORECASE):
          group = "HC"                                                                                                                                                                                                                                       
      elif re.match(r"^C", folder_name, re.IGNORECASE):
          group = "HC"                                                                                                                                                                                                                                       
      elif re.match(r"^PD", folder_name, re.IGNORECASE):
          group = "PD"      
      else:   
          group = "unknown"

      try:                                                                                                                                                                                                                                                   
          df = pd.read_csv(csv_path)
      except Exception as e:                                                                                                                                                                                                                                 
          print(f"  [WARNING] Could not read {csv_path}: {e}")
          continue
                                                                                                                                                                                                                                                             
      df.insert(0, "patient_id", patient_id)
      df.insert(1, "group", group)                                                                                                                                                                                                                           
      df["source_file"] = str(csv_path)
      records.append(df)                                                                                                                                                                                                                                     
      print(f"  [{group}] {patient_id}  ←  {csv_path.relative_to(root_dir)}")
                                                                                                                                                                                                                                                             
  if not records:                                                                                                                                                                                                                                            
      print("No spot_analysis.csv files found.")                                                                                                                                                                                                             
      return pd.DataFrame()                                                                                                                                                                                                                                  
              
  collated = pd.concat(records, ignore_index=True)

  output_path = Path(output_path)                                                                                                                                                                                                                            
  collated.to_hdf(output_path, key="spot_analysis", mode="w", complevel=5, complib="blosc")
  print(f"\nSaved {len(collated):,} rows → {output_path}")                                                                                                                                                                                                   
  print(f"  HCs : {(collated['group']=='HC').sum():,} rows  "                                                                                                                                                                                                
        f"({collated.loc[collated['group']=='HC','patient_id'].nunique()} patients)")                                                                                                                                                                        
  print(f"  PDs : {(collated['group']=='PD').sum():,} rows  "                                                                                                                                                                                                
        f"({collated.loc[collated['group']=='PD','patient_id'].nunique()} patients)")                                                                                                                                                                        
                                                                                                                                                                                                                                                             
  return collated                                                                                                                                                                                                                                            
```

```python
def collate_lo_analysis(root_dir, output_path="collated_data.h5"):
  """                                                                                                                                                                                                                                                        
  Recursively finds all spot_analysis.csv files under root_dir,
  tags each row with patient_id and group (HC or PD), and saves
  the collated DataFrame to an HDF5 file.                                                                                                                                                                                                                    
                                                                                                                                                                                                                                                             
  HC: folder name starts with 'C' or 'PDC'                                                                                                                                                                                                                   
  PD: folder name starts with 'PD' (but not 'PDC')                                                                                                                                                                                                           
                                                                                                                                                                                                                                                             
  Patient ID extraction (from the immediate parent folder name):
    - If folder contains '-': take the substring before the first '-'                                                                                                                                                                                        
    - Otherwise: take the substring before the first '_'                                                                                                                                                                                                     

  Parameters                                                                                                                                                                                                                                                 
  ----------  
  root_dir : str or Path                                                                                                                                                                                                                                     
      Top-level directory to search recursively.
  output_path : str or Path
      Path for the output .h5 file.                                                                                                                                                                                                                          

  Returns                                                                                                                                                                                                                                                    
  -------     
  pd.DataFrame
      The collated DataFrame (also saved to output_path).
  """                                                                                                                                                                                                                                                        
  root_dir = Path(root_dir)
  records = []                                                                                                                                                                                                                                               
                                                                                                                                                                                                                                                             
  for csv_path in sorted(root_dir.rglob("largeobject_analysis.csv")):
      folder_name = csv_path.parent.name                                                                                                                                                                                                                     
                                                                                                                                                                                                                                                             
      # --- determine patient_id ---
      if "-" in folder_name:                                                                                                                                                                                                                                 
          patient_id = folder_name.split("-")[0]
      elif "_" in folder_name:
          patient_id = folder_name.split("_")[0]                                                                                                                                                                                                             
      else:
          patient_id = folder_name                                                                                                                                                                                                                           
              
      # --- determine group ---
      if re.match(r"^PDC", folder_name, re.IGNORECASE):
          group = "HC"                                                                                                                                                                                                                                       
      elif re.match(r"^C", folder_name, re.IGNORECASE):
          group = "HC"                                                                                                                                                                                                                                       
      elif re.match(r"^PD", folder_name, re.IGNORECASE):
          group = "PD"
      elif re.match(r"^negative", folder_name, re.IGNORECASE):
          group = "negative"
      elif re.match(r"^positive", folder_name, re.IGNORECASE):
          group = "positive"
      else:   
          group = "unknown"

      try:                                                                                                                                                                                                                                                   
          df = pd.read_csv(csv_path)
      except Exception as e:                                                                                                                                                                                                                                 
          print(f"  [WARNING] Could not read {csv_path}: {e}")
          continue
                                                                                                                                                                                                                                                             
      df.insert(0, "patient_id", patient_id)
      df.insert(1, "group", group)                                                                                                                                                                                                                           
      df["source_file"] = str(csv_path)
      records.append(df)                                                                                                                                                                                                                                     
      print(f"  [{group}] {patient_id}  ←  {csv_path.relative_to(root_dir)}")
                                                                                                                                                                                                                                                             
  if not records:                                                                                                                                                                                                                                            
      print("No largeobject_analysis.csv files found.")                                                                                                                                                                                                             
      return pd.DataFrame()                                                                                                                                                                                                                                  
              
  collated = pd.concat(records, ignore_index=True)

  output_path = Path(output_path)                                                                                                                                                                                                                            
  collated.to_hdf(output_path, key="largeobject_analysis", mode="w", complevel=5, complib="blosc")
  print(f"\nSaved {len(collated):,} rows → {output_path}")                                                                                                                                                                                                   
  print(f"  HCs : {(collated['group']=='HC').sum():,} rows  "                                                                                                                                                                                                
        f"({collated.loc[collated['group']=='HC','patient_id'].nunique()} patients)")                                                                                                                                                                        
  print(f"  PDs : {(collated['group']=='PD').sum():,} rows  "                                                                                                                                                                                                
        f"({collated.loc[collated['group']=='PD','patient_id'].nunique()} patients)")                                                                                                                                                                        
                                                                                                                                                                                                                                                             
  return collated                                                                                                                                                                                                                                            
```

```python
import os
import tifffile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar


def _lomask_path(image_filename):
    """Derive the loMask tiff path from the raw image path.

    Convention (from RASPRoutines/IOFunctions):
        analysis_dir  = os.path.dirname(image_filename) + '_analysis'
        loMask file   = <base_filename>_loMask.tiff
    """
    img_dir   = os.path.dirname(image_filename)
    base      = os.path.splitext(os.path.basename(image_filename))[0]
    return os.path.join(img_dir + '_analysis', base + '_loMask.tiff')


def plot_aggregate_gallery(
    df_lo,
    N,
    crop_param=20,
    area_range=None,
    photon_range=None,
    seed=None,
    cmap='inferno',
    title=None,
    figsize=None,
):
    """
    Plot an N×N gallery of large-aggregate MIPs with bounding-box crops.

    For each selected aggregate:
      1. Load the saved loMask tiff (same analysis directory as the raw image).
      2. Project the mask to 2D (max over z) and label connected components.
      3. Identify the component whose centroid is closest to the stored (x, y).
      4. Compute its bounding box and expand by crop_param pixels on each side.
      5. Crop the raw image to that region across z-planes [zi, zf] and MIP.

    This gives each panel a crop that is tight to the aggregate's actual footprint
    rather than a fixed window, while the crop_param margin provides consistent context.

    Parameters
    ----------
    df_lo : pd.DataFrame
        Large-object dataframe (e.g. HC_LO or PD_LO). Required columns:
        x, y, zi, zf, area, sum_intensity_in_photons, image_filename.
    N : int
        Size of the square gallery grid (N×N panels).
    crop_param : int
        Extra pixels to add around the bounding box on each side. Default 20.
    area_range : tuple(float, float) or None
        Keep only aggregates with pixel area in [min, max]. None = no filter.
    photon_range : tuple(float, float) or None
        Keep only aggregates with sum_intensity_in_photons in [min, max]. None = no filter.
    seed : int or None
        Random seed for reproducible sampling.
    cmap : str
        Matplotlib colormap. Default 'inferno'.
    title : str or None
        Figure suptitle.
    figsize : tuple(float, float) or None
        Figure size in inches. Auto-computed from N if None.
    """
    rng = np.random.default_rng(seed)
    subset = df_lo.copy().reset_index(drop=True)

    # ── Filter ────────────────────────────────────────────────────────────────
    if area_range is not None:
        subset = subset[
            (subset['area'] >= area_range[0]) & (subset['area'] <= area_range[1])
        ].reset_index(drop=True)
    if photon_range is not None:
        subset = subset[
            (subset['mean_intensity_in_photons'] >= photon_range[0]) &
            (subset['mean_intensity_in_photons'] <= photon_range[1])
        ].reset_index(drop=True)

    if len(subset) == 0:
        raise ValueError('No aggregates remain after filtering — check area_range / photon_range.')

    n_needed = N * N
    if len(subset) < n_needed:
        print(f'Warning: only {len(subset)} aggregates available; '
              f'reducing grid to {int(np.ceil(np.sqrt(len(subset))))}×{int(np.ceil(np.sqrt(len(subset))))}')
        selected = subset
        N = int(np.ceil(np.sqrt(len(selected))))
    else:
        chosen = rng.choice(len(subset), size=n_needed, replace=False)
        selected = subset.iloc[chosen].reset_index(drop=True)

    # ── Image / mask cache ────────────────────────────────────────────────────
    img_cache  = {}
    mask_cache = {}

    def load_image(path):
        if path not in img_cache:
            img_cache[path] = tifffile.imread(path).astype(np.float32)
        return img_cache[path]

    def load_mask(path):
        if path not in mask_cache:
            mask_cache[path] = tifffile.imread(path).astype(bool)
        return mask_cache[path]

    # ── Plot ──────────────────────────────────────────────────────────────────
    if figsize is None:
        figsize = (N * 2.0, N * 2.2)
    fig, axes = plt.subplots(N, N, figsize=figsize)
    axes = np.atleast_1d(axes).reshape(N, N)

    for panel_idx, ax in enumerate(axes.flat):
        if panel_idx >= len(selected):
            ax.axis('off')
            continue

        row      = selected.iloc[panel_idx]
        cx       = int(row['x'])          # centroid row coordinate
        cy       = int(row['y'])          # centroid col coordinate
        img_path = row['image_filename']
        msk_path = _lomask_path(img_path)

        try:
            image = load_image(img_path)
        except Exception as exc:
            ax.text(0.5, 0.5, f'Image load error:\n{exc}', ha='center', va='center',
                    fontsize=4, transform=ax.transAxes)
            ax.axis('off')
            continue

        # ── Get bounding box from loMask ──────────────────────────────────────
        try:
            lo_mask = load_mask(msk_path)
            # Project to 2D (mask may be 3D = one layer per analysed z-plane)
            mask_2d = lo_mask.max(axis=0) if lo_mask.ndim == 3 else lo_mask
            labeled = label(mask_2d)
            props   = regionprops(labeled)
            if len(props) == 0:
                raise ValueError('No labelled regions in mask')
            # Match to the stored centroid (x = row, y = col)
            best = min(props, key=lambda p: (p.centroid[0] - cx)**2 + (p.centroid[1] - cy)**2)
            r0, c0, r1, c1 = best.bbox   # row_min, col_min, row_max, col_max
        except Exception as exc:
            # Fall back to a fixed window if the mask is unavailable
            print(f'  [panel {panel_idx}] mask fallback ({exc}); using fixed ±{crop_param*2} px window')
            r0, r1 = cx - crop_param * 2, cx + crop_param * 2
            c0, c1 = cy - crop_param * 2, cy + crop_param * 2

        # ── Crop raw image to bbox + crop_param margin ────────────────────────
        if image.ndim == 3:
            nz, nr, nc = image.shape
            z0 = max(0, int(row['zi']) - 1)   # zi is 1-indexed
            z1 = min(nz, int(row['zf']))
            rr0 = max(0, r0 - crop_param)
            rr1 = min(nr, r1 + crop_param)
            cc0 = max(0, c0 - crop_param)
            cc1 = min(nc, c1 + crop_param)
            mip = image[z0:z1, rr0:rr1, cc0:cc1].max(axis=0)
        else:
            nr, nc = image.shape
            rr0 = max(0, r0 - crop_param)
            rr1 = min(nr, r1 + crop_param)
            cc0 = max(0, c0 - crop_param)
            cc1 = min(nc, c1 + crop_param)
            mip = image[rr0:rr1, cc0:cc1]

        ax.imshow(mip, cmap=cmap, interpolation='none')
        ax.set_title(
            f"{row['area']:.0f} px²  |  {row['mean_intensity_in_photons']:.0f} ph",
            fontsize=5, pad=1.5
        )
        scalebarlabel='1 µm'
        length_pixels = 1000 / 110
        ax.add_artist(AnchoredSizeBar(
                            ax.transData,
                            length_pixels, scalebarlabel, "lower center",
                            pad=0.5, color="white", frameon=False,
                            size_vertical=length_pixels/20,
                            fontproperties={"size": 6},
                        ))
        ax.axis('off')

    if title:
        fig.suptitle(title, fontsize=10, y=1.01)

    n_used = min(len(selected), N * N)
    print(f'Gallery: {n_used} aggregates shown  '
          f'(filtered pool: {len(subset)},  input: {len(df_lo)} rows)')
    return fig, ax


# ── Example calls ─────────────────────────────────────────────────────────────
# plot_aggregate_gallery(HC_LO, N=5, seed=42, title='HC large aggregates')
# plot_aggregate_gallery(HC_LO, N=4, area_range=(500, 5000), photon_range=(1e4, 1e6), seed=0)
# plot_aggregate_gallery(PD_LO, N=5, crop_param=30, seed=1, title='PD large aggregates')
```

```python
import tifffile
from skimage.measure import label, regionprops
from src import CoincidenceFunctions

C_F = CoincidenceFunctions.Coincidence_Functions()


def _lomask_path(image_filename, protein_string='C561'):
    img_dir = os.path.dirname(image_filename)
    base = os.path.splitext(os.path.basename(image_filename))[0]
    return os.path.join(img_dir + '_analysis', base + '_loMask.tiff')


def _cellmask_path(image_filename, protein_string='C561', cell_string='C488'):
    img_dir = os.path.dirname(image_filename)
    base = os.path.splitext(os.path.basename(image_filename))[0]
    prefix = base.split(protein_string)[0]
    return os.path.join(img_dir + '_analysis', prefix + cell_string + '_cellMask.tiff')


def annotate_lo_incell(df, protein_string='C561', cell_string='C488'):
    """
    For each LO in df, checks whether its 3D pixel footprint overlaps with
    the microglial cell mask.

    Uses skimage.label on the full 3D loMask so each LO (regardless of how
    many z-planes it spans) is a single connected component. Matches each CSV
    row to the closest labelled region by 3D centroid (z is 1-indexed in the
    CSV, 0-indexed in the mask array).

    Returns a boolean numpy array aligned with df after reset_index.
    """
    df = df.reset_index(drop=True)
    in_cell = np.zeros(len(df), dtype=bool)

    for image_filename in np.unique(df['image_filename']):
        img_mask   = df['image_filename'] == image_filename
        row_indices = np.where(img_mask)[0]
        sub_df     = df.loc[img_mask].reset_index(drop=True)

        lo_path   = _lomask_path(image_filename, protein_string)
        cell_path = _cellmask_path(image_filename, protein_string, cell_string)

        if not os.path.isfile(lo_path):
            print(f'  [SKIP] No loMask:   {os.path.basename(lo_path)}')
            continue
        if not os.path.isfile(cell_path):
            print(f'  [SKIP] No cellMask: {os.path.basename(cell_path)}')
            continue

        lo_mask   = tifffile.imread(lo_path).astype(bool)
        cell_mask = tifffile.imread(cell_path).astype(bool)

        # Make both masks the same dimensionality
        if lo_mask.ndim == 3 and cell_mask.ndim == 2:
            cell_mask = np.broadcast_to(cell_mask, lo_mask.shape).copy()
        elif lo_mask.ndim == 2 and cell_mask.ndim == 3:
            lo_mask = np.broadcast_to(lo_mask, cell_mask.shape).copy()

        # Label 3D connected components — one label per LO regardless of z-extent
        labeled = label(lo_mask)
        props   = regionprops(labeled)
        if len(props) == 0:
            continue

        # Flat pixel indices per labelled LO
        image_size = lo_mask.shape
        largeobj_indices = []
        for prop in props:
            c = prop.coords          # (N, ndim): (z,y,x) for 3D, (y,x) for 2D
            if lo_mask.ndim == 3:
                flat = np.ravel_multi_index((c[:, 0], c[:, 1], c[:, 2]), image_size)
            else:
                flat = np.ravel_multi_index((c[:, 0], c[:, 1]), image_size)
            largeobj_indices.append(flat)

        # Flat cell mask indices
        mask_indices = np.ravel_multi_index(np.where(cell_mask), cell_mask.shape)

        # Per-LO overlap boolean
        overlap = C_F.test_largeobj_mask_overlap(
            largeobj_indices, mask_indices, len(props), raw=True
        ).astype(bool)

        # Match each CSV row to the closest labelled region by centroid proximity
        # regionprops centroids are 0-indexed; CSV z is 1-indexed
        centroids = np.array([p.centroid for p in props])  # (n_props, ndim)

        for local_i, row in sub_df.iterrows():
            if lo_mask.ndim == 3:
                cz = float(row['z']) - 1   # 1-indexed → 0-indexed
                cx, cy = float(row['x']), float(row['y'])
                dists = ((centroids[:, 0] - cz)**2 +
                         (centroids[:, 1] - cx)**2 +
                         (centroids[:, 2] - cy)**2)
            else:
                cx, cy = float(row['x']), float(row['y'])
                dists  = (centroids[:, 0] - cx)**2 + (centroids[:, 1] - cy)**2
            in_cell[row_indices[local_i]] = overlap[int(np.argmin(dists))]

    return in_cell
```

```python
kept_HC = kept_HC.copy().reset_index(drop=True)
kept_PD = kept_PD.copy().reset_index(drop=True)

kept_HC['incell'] = annotate_lo_incell(kept_HC)
kept_PD['incell'] = annotate_lo_incell(kept_PD)

print(f"HC: {kept_HC['incell'].sum():.0f} / {len(kept_HC)} LOs in microglia "
      f"({100 * kept_HC['incell'].mean():.1f}%)")
print(f"PD: {kept_PD['incell'].sum():.0f} / {len(kept_PD)} LOs in microglia "
      f"({100 * kept_PD['incell'].mean():.1f}%)")
```

```python
def per_patient_incell_fraction(df):
    rows = []
    for pid in df['patient_id'].unique():
        sub  = df[df['patient_id'] == pid]
        n    = len(sub)
        n_in = int(sub['incell'].sum())
        rows.append({
            'patient_id':       pid,
            'group':            sub['group'].iloc[0],
            'n_lo_total':       n,
            'n_lo_in_cell':     n_in,
            'fraction_in_cell': n_in / n if n > 0 else np.nan,
        })
    return pd.DataFrame(rows)

summary_HC  = per_patient_incell_fraction(kept_HC)
summary_PD  = per_patient_incell_fraction(kept_PD)
summary_all = pd.concat([summary_HC, summary_PD], ignore_index=True)
print(summary_all.to_string(index=False))
```

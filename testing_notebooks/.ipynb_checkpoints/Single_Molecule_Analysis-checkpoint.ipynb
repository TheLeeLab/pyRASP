```python
def FOV_stitcher(folder):
    list_images = np.sort([os.path.join(folder, x) for x in os.listdir(folder) if '.tif' in x and 'Image_Stack' not in x])
    new_image = np.zeros([len(list_images), 1200, 1200])
    for i, file in enumerate(list_images):
        new_image[i, :, :] = IO.read_tiff(file)
    IO.write_tiff(new_image, os.path.join(folder, 'Image_Stack.tif'))
    for file in list_images:
        os.remove(file)
    return new_image
```

```python
for subfolder in subfolders:
    new_image = FOV_stitcher(os.path.join(folder, subfolder))
```

```python
folder = '/home/jbeckwith/Documents/Cambridge University Dropbox/Joseph Beckwith/Chemistry/Lee/Data/Sycamore/20250429_single_Atto565_onSlide/'
subfolders = os.listdir(folder)
```

```python
for subfolder in subfolders:
    data = pl.read_csv(os.path.join(folder, subfolder, 'Puncta_Analysis.csv'))
    molecular_brightnesses = {}
    molecular_ids = np.sort(np.unique(data['molecule_id'].to_numpy()))
    molecular_brightnesses['molecule_id'] = molecular_ids
    sum_photons = np.zeros_like(molecular_ids)
    for i, id_value in enumerate(molecular_ids):
        molecule = data.filter(pl.col("molecule_id") == id_value)
        t = molecule['frame'].to_numpy()
        photons = np.nan_to_num(molecule['sum_intensity_in_photons'].to_numpy())
        
        model = "l2"  # "l1", "rbf", "linear", "normal", "ar",...
        algo = rpt.Binseg(model=model).fit(photons)
        
        sigma = np.std(photons)
        n = len(photons)
        my_bkps = algo.predict(n_bkps=1)
        sum_photons[i] = np.sum(photons[:my_bkps[0]])
    molecular_brightnesses['sum_photons'] = sum_photons
    pl.DataFrame(data=molecular_brightnesses).write_csv(os.path.join(folder, subfolder, 'Photon_Counts.csv'))
```

```python
overall_photons = None
for subfolder in subfolders:
    data = pl.read_csv(os.path.join(folder, subfolder, 'Photon_Counts.csv'))
    if overall_photons is None:
        overall_photons = data
    else:
        overall_photons = pl.concat([overall_photons, data])
```

```python
fig, axs = plotter.one_column_plot()

x = overall_photons['sum_photons'].to_numpy()
axs = plotter.histogram_plot(axs=axs[0], data=x, bins=np.histogram_bin_edges(x, 'fd'), xaxislabel='photons per molecule', density=False)

ymin, ymax = axs.get_ylim()
axs.vlines(x=np.mean(x), ymin=ymin, ymax=ymax, ls='--', color='black', label=r'$\mu$ = '+str(int(np.mean(x)))+', lit. value 19,714 photons')

axs.set_ylim([ymin, ymax])
plt.legend(loc='best', fontsize=6)
axs.set_xlim([0, 100000])
plt.savefig('Photons_per_ATTO565.svg', dpi=600, format='svg')
```

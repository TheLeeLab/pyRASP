```python
import os
import sys

import matplotlib.pyplot as plt
import matplotlib.patches as patches

sys.path.append("..")  # Adds higher directory to python modules path.
from src import RASPRoutines
import polars as pl
import numpy as np
from src import IOFunctions
from src import AnalysisFunctions
from src import Image_Analysis_Functions
from src import HelperFunctions

RASP = RASPRoutines.RASP_Routines()
IO = IOFunctions.IO_Functions()
A_F = AnalysisFunctions.Analysis_Functions()
IA_F = Image_Analysis_Functions.ImageAnalysis_Functions()
H_F = HelperFunctions.Helper_Functions()

from src import PlottingFunctions
plots = PlottingFunctions.Plotter()

import warnings
from copy import copy
import time

warnings.filterwarnings("ignore")
```

```python
folder = '/scratch/sycamore-asap/ASAP_Imaging_Data/Joe/20250227_pilot2BigScan/PD/neuron/patient1'
raw_cell_MIP = IO.read_tiff(os.path.join(folder, 'cell_files', 'cell_bigimage.tiff'))
raw_asyn_MIP = IO.read_tiff(os.path.join(folder, 'asyn_files', 'asyn_bigimage.tiff'))
```

```python
folder = '/scratch/sycamore-asap/ASAP_Imaging_Data/Joe/20250227_pilot2BigScan/PD/neuron/patient1_analysis'
spot_analysis = pl.read_csv(os.path.join(folder, 'spot_analysis.csv'))
threshold = np.percentile(spot_analysis['sum_intensity_in_photons'], 20)
spot_analysis = spot_analysis.filter(pl.col('sum_intensity_in_photons') > threshold)
```

```python
cell_MIP_fast = copy(raw_cell_MIP/np.max(raw_cell_MIP) * 255).astype(np.uint8)
asyn_MIP_fast = copy(raw_asyn_MIP/np.max(raw_asyn_MIP) * 255).astype(np.uint8)
```

```python
import skimage as ski
from skimage.filters import gaussian, threshold_li
from scipy.ndimage import (
    binary_opening,
    binary_closing,
    binary_fill_holes,
    median_filter,
)
```

```python
large_mask = gaussian(cell_MIP_fast, sigma=2, truncate=2.0) - gaussian(
            cell_MIP_fast, sigma=10, truncate=2.0
        )
large_mask = np.abs(large_mask)/np.max(np.abs(large_mask))
large_mask = large_mask > 0.15
plt.imshow(large_mask[5000:7000, 5000:7000])
```

```python
large_mask = binary_opening(large_mask, structure=ski.morphology.disk(1))
large_mask = binary_closing(large_mask, structure=ski.morphology.disk(5))
pixel_index_list, *rest = IA_F.calculate_region_properties(large_mask)
idx1 = np.zeros_like(pixel_index_list, dtype=bool)
imcopy = np.copy(cell_MIP_fast)

for i in np.arange(len(pixel_index_list)):
    idx1[i] = 1 * (
        np.sum(
            imcopy[pixel_index_list[i][:, 0], pixel_index_list[i][:, 1]]
            > 0.05
        )
        / len(pixel_index_list[i][:, 0])
        > 0.1
    )
```

```python
if len(idx1) > 0:
    large_mask = IA_F.create_filled_region(
        cell_MIP_fast.shape, pixel_index_list[idx1]
    )
    large_mask = binary_fill_holes(large_mask)
```

```python
large_mask = ski.morphology.remove_small_holes(
            large_mask, area_threshold=50
        )
large_mask = ski.morphology.remove_small_objects(
    large_mask, min_size=100
)
```

```python
plt.imshow(large_mask[5000:7000, 5000:7000])
```

```python
files = np.unique(spot_analysis['image_filename'].to_numpy())
x_array = []
y_array = []
n_xy = 30
imd = 1200
n_x_range = np.arange(0, imd*n_xy + imd, imd)
filecounter = 0
for j in np.arange(n_xy):
    y = n_x_range[j]
    y_p1 = n_x_range[j+1]
    for i in np.arange(n_xy):
        if j % 2 == 0:
            x = n_x_range[i]
            x_p1 = n_x_range[i+1]
        else:
            x = n_x_range[::-1][i+1]
            x_p1 = n_x_range[::-1][i]
        file = [x for x in files if 'T'+str(filecounter).zfill(3) in x]
        if len(file) > 0:
            file = file[0]
            subset = spot_analysis.filter(pl.col('image_filename') == file)
            x_array = np.hstack([x_array, subset['x'].to_numpy() + x])
            y_array = np.hstack([y_array, subset['y'].to_numpy() + y])
        filecounter += 1
```

```python
fig, axs = plots.two_column_plot(ncolumns=3, widthratio=[1,1,1])

axs[0] = plots.image_plot(axs[0], cell_MIP_fast, scalebarsize=1000*1000, scalebarlabel='1 mm', cbar='off', label='MAP2 stain')

xmin = 5000
xmax = 7000
xsize = 2000

rect = patches.Rectangle((xmin, xmin), xsize, xsize, linewidth=1, edgecolor='white', facecolor='none')

# Add the patch to the Axes
axs[0].add_patch(rect)

axs[1] = plots.image_plot(axs[1], cell_MIP_fast[xmin:xmax, xmin:xmax], scalebarsize=10000, cbar='off', label='MAP2 stain',
                          mask=large_mask[xmin:xmax, xmin:xmax], maskcolor='lightblue', masklinewidth=0.5,
                         plotmask=True)

xmin = 6140
ymin = 6550
size = 240

rect = patches.Rectangle((ymin-5000, xmin-5000), size, size, linewidth=1, edgecolor='white', facecolor='none')

# Add the patch to the Axes
axs[1].add_patch(rect)

selection = (x_array < xmin+size) * (x_array > xmin) * (y_array < ymin+size) * (y_array > ymin)
xdata = x_array[selection]
ydata = y_array[selection]

for i in np.arange(len(xdata)):
    if large_mask[int(xdata[i]), int(ydata[i])] == 0:
        xdata[i] = 0
        ydata[i] = 0

selection = xdata > 0
xdata = xdata[selection]
ydata = ydata[selection]

axs[2] = plots.image_scatter_plot(axs[2], cell_MIP_fast[xmin:xmin+size, ymin:ymin+size], vmin=np.percentile(cell_MIP_fast[xmin:xmin+size, ymin:ymin+size], 1),
                          vmax=np.percentile(cell_MIP_fast[xmin:xmin+size, ymin:ymin+size], 99),
                          scalebarsize=10000, cbar='off', label='MAP2 stain',
                          mask=large_mask[xmin:xmin+size, ymin:ymin+size], maskcolor='lightblue', masklinewidth=1,
                                  xdata=xdata-xmin, ydata=ydata-ymin, s=6, lws=0.5,
                                  facecolor='red',
                         plotmask=True)


xmin_x = 6140
ymin_y = 6550
xmin = xdata[11]-8
ymin = ydata[11]-8
size = 16

rect = patches.Rectangle((ymin-ymin_y, xmin-xmin_x), size, size, linewidth=1, edgecolor='white', facecolor='none')
# Add the patch to the Axes
axs[2].add_patch(rect)

plt.savefig('exemplar_PD_stain.svg', dpi=600, format='svg')
```

```python
fig, axs = plots.one_column_plot()
ROIsize = 8
i = 11
data = raw_asyn_MIP[int(xdata[i]-ROIsize+2):int(xdata[i]+ROIsize+2), int(ydata[i]-ROIsize-1):int(ydata[i]+ROIsize-1)]
axs[0] = plots.image_plot(axs=axs[0], data=data, cbar='off', scalebarlabel='200 nm', scalebarsize=200)
plt.savefig('exemplar_PD_oligomer.svg', dpi=600, format='svg')
plt.show()
```

```python
ROIsize = 8

for i in np.arange(len(xdata)):
    fig, axs = plots.one_column_plot()
    data = raw_asyn_MIP[int(xdata[i]-ROIsize):int(xdata[i]+ROIsize), int(ydata[i]-ROIsize):int(ydata[i]+ROIsize)]
    axs[0] = plots.image_plot(axs=axs[0], data=data, cbar='off', scalebarlabel='200 nm', scalebarsize=200)
    print(i)
    plt.show()
```

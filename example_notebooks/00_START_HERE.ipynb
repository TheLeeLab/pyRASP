## Core pipeline

| Notebook | What it does |
|---|---|
| `Step1_RadialityCalibration.ipynb` | **Run first.** Calibrates the radiality threshold from negative-control images. Saves parameters to disk. Required for any spot detection. |
| `Step2_SingleImageAnalysis.ipynb` | Visual QC on one image: plots detected spots before committing to the full folder run. |
| `Step3_ImageFolderAnalysis.ipynb` | Runs `analyse_images()` on an entire folder. Produces `spot_analysis.csv` and `largeobject_analysis.csv`. |
| `Step4a_LargeObject_CellColocalisation.ipynb` | Re-runs LO-to-cell colocalisation on existing CSVs (uses `lo_colocalise_wrapper()`). Needs Step 3 output. |
| `Step4b_CountPunctaInCells.ipynb` | Counts spots per individual cell (uses `count_puncta_in_individual_cells_threshold_3D()`). Needs Step 3 output. |

## Optional — spatial statistics (needs Step 3 output)

These can be run in any order after Step 3:

| Notebook | What it does |
|---|---|
| `Optional_SpotToSpot_RDF.ipynb` | Auto-correlation RDF: do spots cluster with themselves? |
| `Optional_SpotToMask_RDF.ipynb` | Spot-to-mask RDF: do spots avoid or seek the cell boundary? |
| `Optional_TwoPunctaSets_RDF.ipynb` | Cross-channel RDF: do channel-1 and channel-2 spots co-cluster? |
| `Optional_SpotToSpotCoincidence.ipynb` | Coincidence fraction: what fraction of C1 spots co-localise with C2? |
| `Optional_TwoColourCoincidencePlot.ipynb` | Two-colour coincidence visualisation. **Edit paths before running.** |
| `Optional_CellPunctumPlot.ipynb` | Per-cell punctum analysis plot. |
